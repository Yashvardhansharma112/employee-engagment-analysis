# HR Employee Attrition Analysis and Prediction

## Project Overview

This project aims to analyze the factors contributing to employee attrition at a company and build a predictive model to identify employees at risk of leaving. By understanding the key drivers of attrition, the company can develop targeted strategies to improve employee retention.

The analysis is based on the dataset provided at "/content/WA_Fn-UseC_-HR-Employee-Attrition.csv".

## Data Loading and Initial Inspection

The first step is to load the dataset into a pandas DataFrame and perform an initial inspection to understand its structure and content.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Display the first 5 rows to get a glimpse of the data
print("First 5 rows of the dataset:")
display(df.head())

# Display the shape of the dataset (number of rows and columns)
print("\nShape of the dataset:")
print(df.shape)

# Display data types of each column
print("\nData types of each column:")
print(df.info())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

## Exploratory Data Analysis (EDA)

In this section, we will perform exploratory data analysis to understand the distribution of variables, identify potential relationships, and uncover insights related to employee attrition.

We will start by looking at summary statistics for numerical features and the distribution of categorical features.

```python
# Display descriptive statistics for numerical columns
print("Descriptive statistics for numerical columns:")
display(df.describe())

# Display value counts for categorical columns (showing top 10 for brevity)
print("\nValue counts for categorical columns:")
for col in df.select_dtypes(include='object').columns:
    print(f"\nValue counts for '{col}':")
    print(df[col].value_counts().head(10))
```

### Visualizing Key Distributions

Visualizations help us understand the data distributions more intuitively. We will visualize the distribution of the target variable 'Attrition' and some other key features.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Attrition
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition', data=df)
plt.title('Distribution of Employee Attrition')
plt.show()

# Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Employee Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Distribution of Monthly Income
plt.figure(figsize=(10, 6))
sns.histplot(df['MonthlyIncome'], kde=True)
plt.title('Distribution of Employee Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.show()
```

## Data Preprocessing

Before building a predictive model, we need to preprocess the data. This involves handling categorical variables and splitting the data into training and testing sets.

We will convert categorical features into a numerical format using one-hot encoding and prepare the data for machine learning.

```python
# Convert binary categorical columns to numerical (0 or 1)
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
df['OverTime'] = df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)

# Perform one-hot encoding for other categorical features
df = pd.get_dummies(df, columns=['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus', 'Over18'], drop_first=True)


# Drop irrelevant columns (EmployeeNumber, EmployeeCount, StandardHours have constant values)
df = df.drop(['EmployeeNumber', 'EmployeeCount' , 'StandardHours'] , axis=1)

# Display the first few rows of the processed data
print("First 5 rows of the processed dataset:")
display(df.head())

# Display the shape of the processed dataset
print("\nShape of the processed dataset:")
print(df.shape)
```

## Model Building and Evaluation

Now that the data is preprocessed, we can build a predictive model to forecast employee attrition. We will use a RandomForestClassifier, which is a powerful and commonly used model for classification tasks.

We will split the data into training and testing sets, train the model on the training data, and evaluate its performance on the unseen test data.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Define features (X) and target (y)
x = df.drop('Attrition' , axis=1)
y = df['Attrition']

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y) # Using stratify to maintain attrition distribution

# Initialize and train the RandomForestClassifier model
model = RandomForestClassifier(n_jobs=-1, random_state=42)
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model
print("Model Evaluation:")
print("\nAccuracy:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

### Feature Importance

Let's examine the feature importances from the trained RandomForestClassifier to understand which features were most influential in predicting attrition.

```python
# Display feature importances
if hasattr(model, 'feature_importances_'):
    importances = model.feature_importances_
    feature_names = x_train.columns
    forest_importances = pd.Series(importances, index=feature_names)

    fig, ax = plt.subplots(figsize=(10, 8))
    forest_importances.sort_values(ascending=False).head(10).plot.bar(ax=ax)
    ax.set_title("Top 10 Feature Importances")
    ax.set_ylabel("Mean decrease in impurity")
    plt.tight_layout()
    plt.show()
```
